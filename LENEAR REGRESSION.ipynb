```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,5,13,25,33,46]).reshape((-1,1))
y=np.array([3,6,9,12,14,18])
```

```python
plt.scatter(x,y)
plt.xlabel("D")
plt.ylabel("div")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
SLR=LinearRegression()
```

```python
SLR.fit(x,y)
```

```python
pred=SLR.predict(x)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='Red',marker='*')
print('MSE:',mse)
print("intercept",SLR.intercept_)
print("slope:",SLR.coef_)
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt

x=np.random.randint(100,1000,10).reshape((-1,1))
y=np.random.randint(100,1000,10)
print(x)
print(y)
```

```python
plt.scatter(x,y)
plt.xlabel("D")
plt.ylabel("div")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y,pred)
plt.plot(x,pred,color='blue',marker='*')
print('MSE:',mse)
print("intercept",SLR.intercept_)
print("slope:",SLR.coef_)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
data=pd.read_csv('/content/drive/MyDrive/xy.csv')
x = data["student"].values.reshape(-1,1)
y = data["roll.no"].values
plt.scatter(x,y)
plt.xlabel("student")
plt.ylabel("roll.no")
plt.show()

SLR=LinearRegression()
SLR.fit(x,y)
pred=SLR.predict(x)
mse=mean_squared_error(y,pred)

plt.scatter(x,y)
plt.plot(x,pred,color='blue',marker='*')
plt.xlabel("student")
plt.ylabel("roll.no")
plt.show()

print('MSE:',mse)
print("intercept",SLR.intercept_)
print("slope:",SLR.coef_)
```
